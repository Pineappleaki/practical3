```python
def invert(im, show = True):
    """
    Create an inverted image
    """
    inv_im = 255 - im
    
    if show == True:
        io.imshow(inv_im)
    
    return inv_im
    
    
```

```python
new_inv_im = invert(new_image, show= False)
```

```python
def colour_shift(im, show=True):
    """
    Shifts the colours 1 channel
    """
    
    cs_im = np.zeros_like(im)
    cs_im[:,:,0] = im[:,:,1]
    cs_im[:,:,1] = im[:,:,2]
    cs_im[:,:,2] = im[:,:,0]
    
    if show == True:
        io.imshow(cs_im)
    
    return cs_im
    
        
```

```python
io.imsave ('CS_europe-map.png', colour_shift(image))
```
